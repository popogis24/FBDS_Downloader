```python
import requests

def get_data():
    data = {
        'action': 'download',  # Include the 'action' parameter
        'as': 'APP.tar',
        'type': 'php-tar',
        'baseHref': '/AC/ACRELANDIA/',
        'hrefs': '',
        'hrefs[0]': '/AC/ACRELANDIA/APP/'  # Make sure the 'hrefs[0]' parameter is correctly formatted
    }
    response = requests.post('https://geo.fbds.org.br/AC/ACRELANDIA/', json=data)
    if response.status_code == 200:
        with open('APP.tar', 'wb') as f:
            f.write(response.content)
    print(response.status_code)

get_data()
```
